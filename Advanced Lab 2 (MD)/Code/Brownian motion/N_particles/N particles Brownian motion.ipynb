```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
# Reading the data from a text file
data = open ("positions.txt","r");
pts = data.readlines();

# Creating the x and y arrays
x = [];
y = [];
#z = [];

for line in pts:
    
    #if (line.split()[0]!='x' or line.split()[1]!='y' or line.split()[2]!='z'):
    if (line.split()[0]!='x' or line.split()[1]!='y'):
        x.append(float(line.split()[0]));
        y.append(float(line.split()[1]));
        #z.append(float(line.split()[2]));
```

```python
fig, ax = plt.subplots(1,1,figsize=[6,6]);
ax.set_xlabel('$x$');
ax.set_ylabel('$y$');
ax.set_title('2D Brownian motion');
#ax.set_xlim([0, 1]);
#ax.set_ylim([0, 4.5]);

ax.plot (x, y, color = "goldenrod");
#ax.scatter (x, y, color = "forestgreen", marker = "*");
    
ax.grid();
ax.legend(loc = "lower right");
plt.show(fig); 
```

```python
fig.savefig ("2D_Brownian_motion_1part.jpg")
```
