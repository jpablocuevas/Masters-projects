# Atwood Machine

```python
import sympy as sp
```

parameters

```python
l, m1,m2,g,t=sp.symbols('l, m1,m2,g,t')
x1=sp.Function('x1')(t)
x2=l-x1
```

Kinetic Energy

```python
T=sp.Rational(1,2)*m1*x1.diff(t)**2+sp.Rational(1,2)*m2*x2.diff(t)**2
T.simplify()
```

Potential Energy

```python
U=-m1*g*x1-m2*g*x2
U.simplify()
```

Lagrangian

```python
Lag=(T-U).simplify()
Lag
```

Equations of Motion

```python
def EOM(L,q):
  # This function finds the Euler Lagrange equations of L w.r.t q
  u=sp.diff(L,q)
  v=sp.diff(L,sp.diff(q,t))
  w=sp.diff(v,t)
  return sp.simplify(w-u)
```

```python
EOM(Lag,x1)
```
