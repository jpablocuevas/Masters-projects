```python
import sympy as sp
```

Definition of Quaternion Conjugation and Multiplication

```python
def q_conjugate(q):
  w,x,y,z=q
  return(w,-x,-y,-z)

def q_mult(q1,q2):
  w1,x1,y1,z1=q1
  w2,x2,y2,z2=q2
  w= w1*w2-x1*x2-y1*y2-z1*z2
  x= w1*x2+x1*w2+y1*z2-z1*y2
  y= w1*y2+y1*w2+z1*x2-x1*z2
  z= w1*z2+z1*w2+x1*y2-y1*x2
  return(w,x,y,z)
```

Quaternion rotation

```python
def rotation(vector,axis,angle):
  x,y,z=vector
  nx,ny,nz=axis
  q=(0,x,y,z)
  u=(sp.cos(angle/2),sp.sin(angle/2)*nx,sp.sin(angle/2)*ny,sp.sin(angle/2)*nz)
  q_prime=q_mult(q_mult(u,q),q_conjugate(u))
  return q_prime[1:4]
```

Euler-Rodrigues formula

```python
theta=sp.symbols('theta',positive='True')
x1,x2,x3=sp.symbols('x1,x2,x3')
n1,n2,n3=sp.symbols('n1,n2,n3')
```

```python
r_prime=rotation((x1,x2,x3),(n1,n2,n3),theta)
```

```python
sp.collect(sp.simplify(sp.FU['TR8'](sp.simplify(r_prime[0]).subs(n3**2,1-n1**2-n2**2))),[n1**2,n1*n2,n1*n3])
```

```python
sp.collect(sp.simplify(sp.FU['TR8'](sp.simplify(r_prime[1]).subs(n3**2,1-n1**2-n2**2))),[n2**2,n2*n1,n2*n3])
```

```python
sp.collect(sp.simplify(sp.FU['TR8'](sp.simplify(r_prime[2]).subs(n2**2,1-n1**2-n3**2))),[n3**2,n1*n3,n2*n3])
```
