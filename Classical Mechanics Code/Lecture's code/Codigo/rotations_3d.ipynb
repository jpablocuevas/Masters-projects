```python
import sympy as sp
```

Generators of SO(3)

```python
J1=sp.Matrix([[0,0,0],
              [0,0,-sp.I],
              [0,sp.I,0]])
J2=sp.Matrix([[0,0,sp.I],
              [0,0,0],
              [-sp.I,0,0]])
J3=sp.Matrix([[0,-sp.I,0],
              [sp.I,0,0],
              [0,0,0]])
```

```python
theta=sp.symbols('theta',positive=True)
```

Finite rotation on the x axis:

```python
sp.exp(-sp.I*theta*J1)
```

Finite rotation on the y axis:

```python
sp.exp(-sp.I*theta*J2)
```

Finite rotation on the y axis:

```python
sp.exp(-sp.I*theta*J3)
```

Lie Algebra

```python
def comm(a,b):
  return sp.simplify(a*b-b*a)
```

```python
comm(J1,J2)==sp.I*J3
```

```python
comm(J2,J3)==sp.I*J1
```

```python
comm(J3,J1)==sp.I*J2
```

Isomorphism with SU(2)

```python
from sympy.physics.matrices import msigma
```

```python
comm(msigma(1)/2,msigma(2)/2)==sp.I*msigma(3)/2
```

```python
comm(msigma(2)/2,msigma(3)/2)==sp.I*msigma(1)/2
```

```python
comm(msigma(3)/2,msigma(1)/2)==sp.I*msigma(2)/2
```
