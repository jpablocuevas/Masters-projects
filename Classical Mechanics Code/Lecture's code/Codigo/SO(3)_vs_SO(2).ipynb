```python
import sympy as sp
from sympy.physics.matrices import msigma
```

```python
theta=sp.symbols('theta',positive=True)
```

Most general SU(2) transformation

```python
def U(theta,n):
  return sp.cos(theta/2)*sp.eye(2)-sp.I*(n[0]*msigma(1)+n[1]*msigma(2)+n[2]*msigma(3))*sp.sin(theta/2)
```

We verify that this transformation is the exponentiation of the SU(2) generators

```python
sp.simplify(U(theta,[1,0,0]))==sp.exp(-sp.I*(theta/2)*msigma(1)).simplify()
```

```python
sp.simplify(U(theta,[0,1,0]))==sp.exp(-sp.I*(theta/2)*msigma(2)).simplify()
```

```python
sp.simplify(U(theta,[0,0,1]))==sp.exp(-sp.I*(theta/2)*msigma(3)).simplify()
```

SO(3) transformation built from SU(2)

```python
def R(a,n):
  res=sp.zeros(3)
  for i in range(1,4):
    for j in range (1,4):
      res[i-1,j-1]=sp.simplify(sp.Rational(1,2)*sp.trace(msigma(i)*U(a,n)*msigma(j)*sp.conjugate(sp.transpose(U(a,n)))))
  return res
```

Rotations around the cartesian axes

```python
R(theta,[1,0,0])
```

```python
R(theta,[0,1,0])
```

```python
R(theta,[0,0,1])
```
