# Plot tests

Several plot tests from data imported on text files generated by some code written in C

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

# Circle

```python
# Reading the data from a text file
circle = open ("circle.txt","r");
points_circ=circle.readlines();

# Creating the x and y arrays
x_circ=[];
y_circ=[];

for line in points_circ:
    
    if (line.split()[0]!='x' or line.split()[1]!='y'):
        x_circ.append(float(line.split()[0]));
        y_circ.append(float(line.split()[1]));
```

```python
# Plotting the obtanied 
fig_circ, ax_circ = plt.subplots(1,1);
ax_circ.set_xlabel("$x$");
ax_circ.set_ylabel("$y$");
ax_circ.set_title("Circle");
ax_circ.plot(x_circ,y_circ,color="forestgreen",label="Plot of a circle");
ax_circ.grid();
ax_circ.legend(loc="lower left");
#plt.show(fig_circ);       
```

```python
circle.close();
```

# Sphere

```python
# Reading the data from a text file
sphere = open ("sphere.txt","r");
points_sph=sphere.readlines();

# Creating the x and y arrays
x_sph=[];
y_sph=[];
z_sph=[];

for line in points_sph:
    
    if (line.split()[0]!='x' or line.split()[1]!='y' or line.split()[2]!='z'):
        x_sph.append(float(line.split()[0]));
        y_sph.append(float(line.split()[1]));
        z_sph.append(float(line.split()[2]));
```

```python
# Plotting the sphere
fig_sph = plt.figure();
ax_sph = fig_sph.add_subplot(projection='3d');
for i in range (0,len(x_sph),1):
    ax_sph.scatter(x_sph[i], y_sph[i], z_sph[i], marker="_");


ax_sph.set_xlabel('x');
ax_sph.set_ylabel('y');
ax_sph.set_zlabel('z');
plt.show()
```

```python
sphere.close()
```
