# Plots for the Yee algorithm using an additive source and ABC (absorbing boundary conditions)

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Function to plot the electric field based on the data from a text file
def data_plot (name, num):
    
    # Reading the data from a text file
    data_ptr = open (name + ".txt","r");
    data=data_ptr.readlines();

    # Creating the x and y arrays
    Ez=[];
    x=[];

    for line in data:

        if (line.split()[0]!='Ez' or line.split()[1]!='x'):
            Ez.append(float(line.split()[0]));
            x.append(float(line.split()[1]));
    
    # Plotting the obtanied data 
    
    fig, ax_plot = plt.subplots(1,1);
    ax_plot.set_xlabel("$x$");
    ax_plot.set_ylabel("$Ez[50] (V/m)$");
    ax_plot.set_title("Additive Gaussian source with ABC");
    ax_plot.plot(x,Ez,color="darkblue",label="Electric field pulse");
    ax_plot.grid();
    ax_plot.legend(loc="lower left");
    
    fig.savefig("ABC"+str(num)+".png");
```

```python
# List that holds the names of the files 
data_names = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 's', 'r', 't'];
```

```python
# Loop to generate all the pictures 

i = 0;

for letter in data_names:
    
    i = i+1;
    
    data_plot (letter, i);
```
