```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating an array of points
x=np.linspace(-10, 10, 200, endpoint=True, dtype=float);
```

```python
# Creating the corresponding wave function up to n terms
def wave_func(n,x,a):
    
    psi=[];
    dpsi=0;
    
    for i in x:
        
        if (n==1):
            dpsi=8*np.sqrt(15)/pow(n*np.pi,3)*np.sqrt(2/a)*np.sin(n*np.pi/a*i);
        else:
            for j in range (1,n,2):
                dpsi=dpsi+8*np.sqrt(15)/pow(j*np.pi,3)*np.sqrt(2/a)*np.sin(j*np.pi/a*i);
                
        psi.append(dpsi);
        dpsi=0;
    
    return psi;
```

```python
# Infinite well size along the x-axis
a=3;

# Number of terms for the wave function
n=5;
```

```python
# Wave function
psi=wave_func(n,x,a);

# Original function
f=[];

for i in x:
    f.append(i*(a-i)*np.sqrt(30/pow(a,5)));
```

```python
# Plotting the wave function
fig, ax = plt.subplots(1,1,figsize=[6,6]);
ax.set_xlabel("$x$");
ax.set_ylabel("$\psi(x)$, n = "+str(n));
ax.set_title(r"Potential plot with $\psi(x)$");
ax.set(xlim=(-3, 5), ylim=(-2, 1));

ax.plot(x,psi,color="darkblue",label="Wave function plot");
ax.plot(x,f,color="darkred",label="r$f(x)=\sqrt{\dfrac{30}{a^5}}x(a-x)$");
ax.grid()
ax.legend(loc="lower left");
plt.show(fig);
```
