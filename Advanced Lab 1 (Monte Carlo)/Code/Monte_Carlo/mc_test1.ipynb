```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
# Reading the data from a text file
data = open ("positions.txt","r");
pts = data.readlines();

# Creating the x and y arrays
x =[];
y =[];
z =[];

for line in pts:
    
    #if (line.split()[0]!='x' or line.split()[1]!='y' or line.split()[2]!='z'):
    if (line.split()[0]!='x' or line.split()[1]!='y'):
        x.append(float(line.split()[0]));
        y.append(float(line.split()[1]));
        z.append(float(line.split()[2]));
```

```python
plt.scatter (x,y)
```

```python
fig = plt.figure();
ax = fig.add_subplot(projection='3d');

for i in range (0,len(x),1):
    
    ax.scatter(x[i], y[i], z[i], marker="o", color = "darkblue");

ax.set_xlabel('x');
ax.set_ylabel('y');
ax.set_zlabel('z');
plt.show()
```
